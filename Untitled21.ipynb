tokenize

```python
def tokenize_sentences(sentences):
    words = []
    for sentence in sentences:
        w = extract_words(sentence)
        words.extend(w)
        
    words = sorted(list(set(words)))
    return words
```

now bag of words count frequency of words

```python
def bagofwords(sentence, words, a):
    sentence_words = extract_words(sentence)
    # frequency word count
    for sw in sentence_words:
        for i,word in enumerate(words):
            if word == sw: 
                a[i] += 1
                
    return np.array(a)
```

# fit function

Create a dictionary, fill the value of dictonary, return a dictonary and use it to predict classes

```python
def fit_data(df):
    dict = {}
    class_values = set(df["Y"])
    for i in class_values:
        dict[i] = {}
        curr_class_row = (i == df["Y"])
        data = df[curr_class_row]

        for j in features:
            dict[i][j] = data[j].sum()

        dict[i]["total"] = 0
        for k in features:
            dict[i]["total"] += dict[i][k]
    
    return dict
```

Find probability of each class using dictonary result

```python
n_f = len(features)
def probability(dictionary, data_p, curr_class):
    wrd = extract_words_data(data_p)
    output = np.log((curr_class == df["Y"]).sum())-np.log(11314)
    for word in wrd:
        for k in features:
            if(k == word):
                num = dictionary[curr_class][word] + 1
                den = dictionary[curr_class]["total"] + n_f
                curr_class_value = np.log(num) - np.log(den)
                output = output + curr_class_value
    return output
            
    
```

Store the class which give high value of probability 

```python
def predictSinglePoint(result, data_p):
    classes = result.keys()
    best_p = -1000
    best_class = -1
    for curr_class in classes:
        p_current_class = probability(result, data_p, curr_class)
        if(p_current_class > best_p):
            best_p = p_current_class
            best_class = curr_class
    return best_class    
```

Find the corresponding class 

```python
def predict(result, test_data):
    y_pred = []
    for i in range(len(test_data)):
        pred_class = predictSinglePoint(result, test_data[i])
        y_pred.append(pred_class)
    return y_pred
```
